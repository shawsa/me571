```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import subprocess
import scipy as sp
import scipy.sparse
import scipy.sparse.linalg
import timeit
```

```python
%%bash
mpicc demo_util.c p2.c -o p2 -lm
mpirun -n 16 ./p2 -p 17 > results.csv
head results.csv
```

```python
%%capture
#generate data
data = {}
columns = ['Processors', 'Time', 'Error']

for procs in [1, 2, 4, 8, 16]:
    df = pd.DataFrame(columns=columns)
    for power in range(6, 29):
        file = open("results.csv", "w")
        tr = %timeit -n 1 -r 1 -o w = p = subprocess.run(['mpirun','-n', str(procs), './p2','-p',str(power)],stdout=file)
        file.close()
        error = np.genfromtxt('results.csv', delimiter=',')
        data_row = {'Processors': procs, 'Time': tr.best, 'Error':error}
        df = df.append(pd.Series(data=data_row, name=2**power-1))
    data['p' + '{:02d}'.format(procs) ] = df
panel = pd.Panel(data)
```

# Error Results

```python
errors = np.abs(panel.minor_xs('Error')['p01'])
Ns = [2**p - 1 for p in range(6, 29)]

plt.loglog(Ns, errors, 'b-')
plt.xlabel('N')
plt.ylabel('Error')
```

This shows that we have convergence until we get to an error of arround $10^{-13}$ at which point we level off.

```python
times = np.diag(panel.minor_xs('Time'),-5)
procs = [2**(i) for i in range(len(times))]

plt.loglog(procs, times, 'bo')

# Plot best-fit speed-up line
t_weak = times

plt.loglog(procs,[t_weak[0] for p in procs],'r--')

plt.title('Weak scaling', fontsize=18)
plt.xlabel('Cores')
plt.ylabel("Time (s)")
plt.legend(['Time (s)','Perfect scaling'])
```

This shows that we have fairly good weak scaling.

```python
times = np.array(panel.minor_xs('Time'))[-1]
procs = [2**(i) for i in range(len(times))]

plt.loglog(procs, times, 'bo')
t_strong = times
c = np.polyfit( np.log(procs[:-1]), [np.log(t) for t in times[:-1]],1)
plt.loglog(procs,np.exp(np.polyval(c,np.log(procs))),'r--')

plt.legend(['Time (slope={:6.2f})'.format(c[0]),'Speed-up'])
plt.title('Speed-up',fontsize=18);
plt.xlabel('Processors')
plt.ylabel('Time')
```

This shows that we have quite good strong scaling.
